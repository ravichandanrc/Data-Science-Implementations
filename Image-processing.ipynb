### Gray-scaled image

```python
#receiving image
myImage = matplotlib.pyplot.imread("F:/1AMOD/Courses/Data Science with Python 5460/Lab 5 - image processing/flower.png")
height=myImage.shape[0]
width=myImage.shape[1]

#converting to greyscale
for x in range(0, height-1):
    for y in range(0,width-1):
        r=myImage[x,y,2]
        g=myImage[x,y,1]
        b=myImage[x,y,0] 
	#weights for rgb values
        grey=b*0.114+g*0.587+r*0.2989
        #greyscale image
        myImage[x,y]=grey

#visualizing image
imgplot = matplotlib.pyplot.imshow(myImage,cmap='gray')
matplotlib.pyplot.show()
```
