### Importing the necessary modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import data from pydataset package
from pydataset import data
```

### Loading iris dataset into a variable

```python
irisx = data('iris')
print(irisx.head(5))
irisxy = irisx[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']]
```

## Sub-plots

```python
# Adding the subplot at the specified grid position

ax1 = plt.subplot(2,2,1)
irisx['Species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(8,8),title='Iris Species %',wedgeprops={'linewidth': 0.5, 'edgecolor': 'white'},
    textprops={'size': 'x-large'})
ax1.set_title('Plot 1 - Pie chart')

ax2 = plt.subplot(2,2,2)
plt.imshow(irisxy.corr())
plt.xticks(range(len(irisxy.corr().columns)), irisxy.corr().columns)
plt.yticks(range(len(irisxy.corr().columns)), irisxy.corr().columns)
ax2.set_title('Plot 2 - Heat map using imshow()')

ax3 = plt.subplot(2,2,3)
ax3.set_title('Plot 3 - Scatter plot for 1&2 variables')

plt.scatter(irisx.iloc[:,1], irisx.iloc[:,2], s= 80, marker='x', c='blue')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')


ax4 = plt.subplot(2,2,4)
ax4.set_title('Plot 4 - Scatter plot for 3&4 variables')

plt.scatter(irisx.iloc[:,3], irisx.iloc[:,4], s= 80, marker='x', c='red')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')


plt.suptitle(' Visualizations: Creating Sub-plots')
plt.show()
```
